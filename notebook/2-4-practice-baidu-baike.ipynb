# 实践：抓取百度百科
**在这里，我们构建了一个爬虫程序，以从此[页面](https://baike.baidu.com/item/网络爬虫/5162711)开始抓取百度百科的内容。我们保存一个已经访问过的历史网页，以便持续跟踪。**

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
import random

base_url = "https://baike.baidu.com"
his = ["/item/%E7%BD%91%E7%BB%9C%E7%88%AC%E8%99%AB/5162711"]
```

**选择 "his" 中的最后一个子网址，打印标题和网址。**

```python
url = base_url + his[-1]

html = urlopen(url).read().decode('utf-8')
soup = BeautifulSoup(html, features='lxml')
print(soup.find('h1').get_text(), '    url: ', his[-1])
```

**找到百度百科的所有sub_urls（物品页面），随机选择一个sub_urls并将其存储在“his”中。如果没有找到有效的子链接，则弹出“his”中的最后一个 url。**

```python
# 修正后的正则表达式（关键：移除 $ 并优化模式）
pattern = re.compile(r"^/item/(%[\da-fA-F]{2})+$")  # 更健壮的匹配
sub_urls = soup.find_all("a", {"target": "_blank", "href": re.compile(r"/item/.*%")})
print(sub_urls)
if len(sub_urls) != 0:
    his.append(random.sample(sub_urls, 1)[0]['href'])
else:
    # no valid sub link found
    his.pop()
print(his)
```

**将所有内容整合在一起。进行20次随机运行。看看我们最终会得到什么。**

```python
his = ["/item/%E7%BD%91%E7%BB%9C%E7%88%AC%E8%99%AB/5162711"]

for i in range(20):
    url = base_url + his[-1]

    html = urlopen(url).read().decode('utf-8')
    soup = BeautifulSoup(html, features='lxml')
    print(i, soup.find('h1').get_text(), '    url: ', his[-1])

    # find valid urls
    sub_urls = soup.find_all("a", {"target": "_blank", "href": re.compile(r"/item/.*%")})

    if len(sub_urls) != 0:
        his.append(random.sample(sub_urls, 1)[0]['href'])
    else:
        # no valid sub link found
        his.pop()
```
